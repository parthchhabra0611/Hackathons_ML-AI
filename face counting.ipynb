```python
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from mtcnn import MTCNN
```

```python
# draw an image with detected objects
def draw_image_with_boxes(filename, result_list):

    data = pyplot.imread(filename)
    pyplot.imshow(data)

    ax = pyplot.gca()

    for result in result_list:
        x, y, width, height = result['box']
        rect = Rectangle((x, y), width, height, fill=False, color='red')
        ax.add_patch(rect)

    # show the plot
    pyplot.show()
 
```

../input/training-zip-images/image_data

```python
detector = MTCNN()
```

```python
pixels1 = pyplot.imread('../input/training-zip-images/image_data/10846.jpg')
faces1 = detector.detect_faces(pixels1)
print(len(faces1))
draw_image_with_boxes('../input/training-zip-images/image_data/10846.jpg', faces1)
```
